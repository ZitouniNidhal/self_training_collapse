# Run Chain Examples
This notebook shows how to execute the repository's chain workflows using the provided scripts.

The cells below set the repository root and run the CARE, ES, and GRPO chains with a small example configuration.


```python
from pathlib import Path
import os
import subprocess

repo_root = Path.cwd()
if repo_root.name == 'notebooks':
    repo_root = repo_root.parent
os.chdir(repo_root)
print('Repository root set to:', repo_root)

# Run the CARE chain example
result = subprocess.run([
    'python',
    'scripts/run_care_chain.py',
    '--campaigns', '1',
    '--steps_per_campaign', '5',
    '--output_dir', 'outputs/notebook_care_chain',
], check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)

# Run the ES chain example
result = subprocess.run([
    'python',
    'scripts/run_es_chain.py',
    '--campaigns', '1',
    '--steps_per_campaign', '5',
    '--output_dir', 'outputs/notebook_es_chain',
], check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)

# Run the GRPO chain example
result = subprocess.run([
    'python',
    'scripts/run_grpo_chain.py',
    '--campaigns', '1',
    '--steps_per_campaign', '5',
    '--output_dir', 'outputs/notebook_grpo_chain',
], check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
```
